```python
import pandas as pd
import numpy as np

#import warnings
#warnings.filterwarnings('ignore')

from collections import Counter
```

```python
df_train = pd.read_csv("train.csv")
print(df_train.shape)
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
print(df_test.shape)
df_test.head()
```

```python
df = df_train.append(df_test,sort=False)
df.shape
```

```python
print("NA Values\n")
for col in df.columns:
    print("{} : {}".format(col,len(df[df[col].isna()])))
```

```python
%matplotlib inline
from matplotlib import pyplot as plt,style
style.use("fivethirtyeight")
```

```python
"""
Lets understand following columns
MSSubClass: The building class
MSZoning: The general zoning classification
"""
print("MSSubClass: {}".format(Counter(df.MSSubClass)))
print("MSZoning: {}".format(Counter(df.MSZoning)))

"""
These are important features
Lets encode them
"""
from sklearn.preprocessing import LabelEncoder
df.MSZoning.fillna('RL',inplace=True)

le = LabelEncoder()
df.MSSubClass = le.fit_transform(df.MSSubClass)
df.MSZoning = le.fit_transform(df.MSZoning)
```

```python
"""
LotFrontage: Linear feet of street connected to property
LotArea: Lot size in square feet
"""
print("Mean : {}, Max : {}, Min : {}, Std : {}".format(
    np.mean(df.LotFrontage),np.max(df.LotFrontage),np.min(df.LotFrontage),np.std(df.LotFrontage)))
df.LotFrontage.fillna(np.mean(df.LotFrontage),inplace=True)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.LotFrontage = scaler.fit_transform(np.array(df.LotFrontage).reshape(-1,1)).reshape(1,-1)[0]

print("Mean : {}, Max : {}, Min : {}, Std : {}".format(
    np.mean(df.LotArea),np.max(df.LotArea),np.min(df.LotArea),np.std(df.LotArea)))
df.LotArea.fillna(np.mean(df.LotArea),inplace=True)
df.LotArea = scaler.fit_transform(np.array(df.LotArea).reshape(-1,1)).reshape(1,-1)[0]
```

```python
"""
Street: Type of road access
Alley: Type of alley access
LotShape: General shape of property
LandContour: Flatness of the property
Utilities: Type of utilities available
LotConfig: Lot configuration
LandSlope: Slope of property
Neighborhood: Physical locations within Ames city limits
"""
print("Alley: {}".format(Counter(df.Alley)))
df.Alley.fillna("NA",inplace=True)
df.Alley = le.fit_transform(df.Alley)
df.Street = le.fit_transform(df.Street)
df.LotShape = le.fit_transform(df.LotShape)
df.LandContour = le.fit_transform(df.LandContour)
print("Utilities: {}".format(Counter(df.Utilities)))
df.Utilities.fillna("AllPub",inplace=True)
df.Utilities = le.fit_transform(df.Utilities)
df.LotConfig = le.fit_transform(df.LotConfig)
df.LandSlope = le.fit_transform(df.LandSlope)
df.Neighborhood = le.fit_transform(df.Neighborhood)
```

```python
"""
Condition1: Proximity to main road or railroad
Condition2: Proximity to main road or railroad (if a second is present)
BldgType: Type of dwelling
HouseStyle: Style of dwelling
OverallQual: Overall material and finish quality
OverallCond: Overall condition rating
"""
df.Condition1 = le.fit_transform(df.Condition1)
df.Condition2 = le.fit_transform(df.Condition2)
df.BldgType = le.fit_transform(df.BldgType)
df.HouseStyle = le.fit_transform(df.HouseStyle)
df.OverallCond = le.fit_transform(df.OverallCond)
df.OverallQual = le.fit_transform(df.OverallQual)
```
