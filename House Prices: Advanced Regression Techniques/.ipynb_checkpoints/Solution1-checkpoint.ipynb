```python
import pandas as pd
import numpy as np

#import warnings
#warnings.filterwarnings('ignore')

from collections import Counter
```

```python
df_train = pd.read_csv("train.csv")
print(df_train.shape)
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
print(df_test.shape)
df_test.head()
```

```python
df = df_train.append(df_test,sort=False)
df.shape
```

```python
#print("NA Values\n")
#for col in df.columns:
#    print("{} : {}".format(col,len(df[df[col].isna()])))
```

```python
#lets fill na values
from sklearn.preprocessing import LabelEncoder
df.MSZoning.fillna('RL',inplace=True)

df.LotFrontage = df.LotFrontage.astype('float')
df.LotArea = df.LotArea.astype('float')

#LotFrontage
print("Mean : {}, Max : {}, Min : {}, Std : {}".format(
    np.mean(df.LotFrontage),np.max(df.LotFrontage),np.min(df.LotFrontage),np.std(df.LotFrontage)))
df.LotFrontage.fillna(np.mean(df.LotFrontage),inplace=True)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.LotFrontage = scaler.fit_transform(np.array(df.LotFrontage).reshape(-1,1)).reshape(1,-1)[0]

#LotArea
print("Mean : {}, Max : {}, Min : {}, Std : {}".format(
    np.mean(df.LotArea),np.max(df.LotArea),np.min(df.LotArea),np.std(df.LotArea)))
df.LotArea.fillna(np.mean(df.LotArea),inplace=True)
df.LotArea = scaler.fit_transform(np.array(df.LotArea).reshape(-1,1)).reshape(1,-1)[0]
```

```python
#Alley
print("Alley: {}".format(Counter(df.Alley)))
df.Alley.fillna("NA",inplace=True)

#Utilities
print("Utilities: {}".format(Counter(df.Utilities)))
df.Utilities.fillna("AllPub",inplace=True)
```

```python
#YearBuilt and YearRemodAdd to continuos variables
df.YearBuilt = df.YearBuilt - np.min(df.YearBuilt)
df.YearBuilt = df.YearBuilt.astype('float')
df.YearRemodAdd = df.YearRemodAdd - np.min(df.YearRemodAdd)

df.YearRemodAdd = df.YearRemodAdd.astype('float')
df.YearBuilt = scaler.fit_transform(np.array(df.YearBuilt).reshape(-1,1)).reshape(1,-1)[0]
df.YearRemodAdd = scaler.fit_transform(np.array(df.YearRemodAdd).reshape(-1,1)).reshape(1,-1)[0]
```

```python
"""
Exterior1st,Exterior2nd,MasVnrType,MasVnrArea
"""
print("Exterior1st: {}".format(Counter(df.Exterior1st)))
print("Exterior2nd: {}".format(Counter(df.Exterior2nd)))
print("MasVnrType: {}".format(Counter(df.MasVnrType)))
print("MasVnrArea : Mean : {}, Max : {}, Min : {}, Std : {}".format(
    np.mean(df.MasVnrArea),np.max(df.MasVnrArea),np.min(df.MasVnrArea),np.std(df.MasVnrArea)))


df.Exterior1st.fillna('VinylSd',inplace=True)
df.Exterior2nd.fillna('VinylSd',inplace=True)
df.MasVnrType.fillna('None',inplace=True)
df.MasVnrArea.fillna(np.mean(df.MasVnrArea),inplace=True)
```

```python
"""
BsmtQual,BsmtCond,BsmtExposure,BsmtFinType1
Nan means no basement
"""
print("BsmtQual: {}".format(Counter(df.BsmtQual)))
print("BsmtCond: {}".format(Counter(df.BsmtCond)))
print("BsmtExposure: {}".format(Counter(df.BsmtExposure)))
print("BsmtFinType1: {}".format(Counter(df.BsmtFinType1)))

df.BsmtQual.fillna('NA',inplace=True)
df.BsmtCond.fillna('NA',inplace=True)
df.BsmtExposure.fillna('NA',inplace=True)
df.BsmtFinType1.fillna('NA',inplace=True)
```
