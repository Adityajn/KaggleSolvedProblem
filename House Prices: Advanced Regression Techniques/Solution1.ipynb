```python
categorical_features=["MSSubClass","MSZoning","Street","Alley","LotShape","LandContour","Utilities","LotConfig",
                        "LandSlope","Neighborhood","Condition1","Condition2","BldgType","HouseStyle","OverallQual",
                      "OverallCond","RoofStyle","RoofMatl","Exterior1st","Exterior2nd","MasVnrType","ExterQual",
                     "ExterCond","Foundation","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2",
                     "Heating","HeatingQC","CentralAir","Electrical","KitchenQual","Functional","FireplaceQu",
                      "GarageType","GarageFinish","GarageQual","GarageCond","GarageCars","PavedDrive","PoolQC",
                      "Fence","MiscFeature","SaleType","SaleCondition"]
```

```python
train = df[:len(df_train)]
test = df[len(df_train):]

Y = train.SalePrice
X = train.drop(['SalePrice','Id'],axis=1)

X_test_id = test['Id']
X_test = test.drop(['SalePrice','Id'],axis=1)

from sklearn.model_selection import train_test_split
X_train,X_validation, Y_train,Y_validation = train_test_split(X,Y,test_size=0.2)
(X_train.shape,X_test.shape,X_validation.shape)
```

```python
print(len(X_train.columns),len(categorical_features))
categorical_feature_indices= []
for cat_feature in categorical_features:
    categorical_feature_indices.append(list(X_train.columns).index(cat_feature))
len(categorical_feature_indices)
```

```python
from catboost import CatBoostRegressor,Pool
from sklearn.metrics import mean_squared_error
train_pool = Pool(X_train,Y_train,cat_features=categorical_feature_indices)
validation_pool = Pool(X_validation,Y_validation,cat_features=categorical_feature_indices)
```

```python
params = {
    'iterations': 500,
    'learning_rate': 0.1,
    'eval_metric': 'Accuracy',
    'random_seed': 42,
    'logging_level': 'Verbose',
    'use_best_model': True
}
model = CatBoostRegressor()
model.fit(train_pool, eval_set=validation_pool);
print('Best model validation accuracy: {:.4}'.format(
    mean_squared_error(Y_validation, model.predict(X_validation))
))
model.score(validation_pool)
```
