```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
df.head(5)
```

##### datetime         hourly date + timestamp  
##### season           1 = spring, 2 = summer, 3 = fall, 4 = winter 
##### holiday           whether the day is considered a holiday
##### workingday    whether the day is neither a weekend nor holiday
##### weather          1: Clear, Few clouds, Partly cloudy, Partly cloudy
#####                         2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
#####                         3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
#####                         4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 
##### temp                temperature in Celsius
##### atemp             "feels like" temperature in Celsius
##### humidity         relative humidity
##### windspeed     wind speed
##### casual             number of non-registered user rentals initiated
##### registered       number of registered user rentals initiated
##### count              number of total rentals

```python
datetime = df['datetime'].str.split(' ')
df.head(5)
```

```python
df['timepart'] = [ int(int(i[1].split(':')[0])/4) for i in datetime ]
df.__delitem__('datetime')
```

```python
df_y = df['count']
df_train=df[['season','holiday','workingday','temp','atemp','humidity','windspeed','casual','registered']]
df_train.head(2)
```

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor()
model.fit(df_train,df_y)
model.score(df_train,df_y)
```

```python
df2 = pd.read_csv('test.csv')
df2_id = df2['datetime']
datetime2 = df2['datetime'].str.split(' ')
df2['timepart'] = [ int(int(i[1].split(':')[0])/4) for i in datetime2 ]
df2.__delitem__('datetime')
df2.head(5)
```

```python
df2_test = df2
df2_predict = model.predict(df2)
len(df2_predict)==len(df2_id)
df2_predict = np.array( [round(i) for i in df2_predict])
```

```python
res = np.column_stack((df2_id,df2_predict))
ans = pd.DataFrame(data=res,columns=['datetime','count'])
ans.to_csv('sol.csv',index=False)
```
